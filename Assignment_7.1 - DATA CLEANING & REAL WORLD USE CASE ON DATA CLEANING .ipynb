```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data =  pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
label = pd.DataFrame(list(map(lambda x: iris.target_names[x],iris.target)), columns = ['Species'])
```

```python
iris = pd.concat([data,label],axis=1)
```

```python
print (iris.head())
```

```python
# Distplot
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
sns.distplot(iris['sepal length (cm)'], color='r')
plt.subplot(2,2,2)
sns.distplot(iris['sepal width (cm)'], color= 'y')
plt.subplot(2,2,3)
sns.distplot(iris['petal length (cm)'], color= 'b')
plt.subplot(2,2,4)
sns.distplot(iris['petal length (cm)'], color= 'g')

```
